```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from load_data import *
from train import *
```

```python
def eda(dataset_dir):
    train_dataset = load_data(dataset_dir)
    train_dataset = train_dataset[0]
    train_label = label_to_num(train_dataset['label'].values)
    train_dataset['label'] = train_label

    cnts = train_dataset['label'].value_counts()
    cnts = pd.DataFrame(cnts)
    cnts = cnts.sort_index()
    label_cnts = cnts['label'].to_list()

    x = np.arange(0, 10)
    y = label_cnts

    plt.figure(figsize= (6, 6))
    sns.set_style('whitegrid')
    sns.countplot(x='label', data=train_dataset)

    for i, v in enumerate(x):
        plt.text(v, y[i], y[i],                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
                fontsize = 9, 
                horizontalalignment='center',  # horizontalalignment (left, center, right)
                verticalalignment='bottom')    # verticalalignment (top, center, bottom)
    plt.show()
```

```python
# 전체 데이터셋
dataset_dir = "ruw_data/merged_dataset.csv"
pd_dataset = pd.read_csv(dataset_dir)
print('전체 데이터 개수: ', len(pd_dataset))
print('결측치:\n', pd_dataset.isnull().sum())

eda(dataset_dir)
```
